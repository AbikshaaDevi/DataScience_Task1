```python
x = df.drop(['Final year Result','Extracurricular Activities','Distance (km)'], axis=1)
y = df['Final year Result']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

x['Internet availability at home'] = le.fit_transform(x['Internet availability at home'])
y=le.fit_transform(y)
```

```python
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
input = input.reindex(columns=x.columns, fill_value=0)
```

```python
a = model.predict(input)
for res in a:
  if(res==1):
    print('Fail')
  elif(res==0):
    print('Distinction')
  else:
    print('Pass')
```
